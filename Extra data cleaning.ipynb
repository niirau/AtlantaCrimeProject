```python
import pandas as pd
propertysales = pd.read_csv('./datasets/med_sale_price.csv', sep='\t')
propertysales = propertysales.T
propertysales.reset_index(inplace=True)
del propertysales['index']
propertysales = propertysales.T
```

```python
propertysales.columns = propertysales.iloc[0]
propertysales = propertysales[1:]
propertysales = pd.melt(propertysales, id_vars=['Region'],var_name='Month')
```

```python
propertysales = propertysales.rename(columns={"value": "Median price"})
propertysales['Region'] = propertysales['Region'].str.replace('Atlanta, GA - ','')

propertysales['Year'] = propertysales['Month'].str.extract('.{1,20}([0-9]{4})')
propertysales['Month'] = propertysales['Month'].str.extract('(.{1,20}) [0-9]{4}')
propertysales.fillna(value='null', inplace=True)

propertysales['Median price'] = propertysales['Median price'].str.replace('K', '000')
propertysales['Median price'] = propertysales['Median price'].str.replace('$', '', regex=False)
propertysales['Median price'] = propertysales['Median price'].str.replace(',','')


propertysales.index.name = 'SalesKey'
propertysales.index.astype('int32')
propertysales
```

```python
months = {1:'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}
propertysales['MonthNo'] = propertysales['Month']
for x in months.keys():
    propertysales['MonthNo'] = propertysales['MonthNo'].str.replace(months[x], str(x))
propertysales
```

```python
propertysales.to_csv('./datasets/med_sale_price_clean.csv', sep='\t')
```
