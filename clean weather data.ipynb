```python
import pandas as pd

df = pd.read_csv('./datasets/weather_description.csv')
df
```

```python
df2 = df[['datetime', 'Atlanta']]

temp = pd.read_csv('./datasets/temperature.csv')
temp = temp[['datetime', 'Atlanta']]
temp['datetime'] = temp['datetime'].astype('datetime64')
condition1 = temp['datetime'].dt.hour == 12
temp = temp[condition1]
temp['datetime'] = temp['datetime'].dt.date
temp = temp.rename(columns={'Atlanta': 'temperature (C)'})

temp
```

```python
df2['datetime'] = df2['datetime'].astype('datetime64')
```

```python
# get only midday values and set datetime to only include date
condition = df2['datetime'].dt.hour == 12
df2 = df2[condition]
df2
```

```python
df2 = df2.rename(columns={'Atlanta': 'description'})
df2['datetime'] = df2['datetime'].dt.date
df2
```

```python
final = df2.merge(temp, how="inner", on=["datetime"])
```

```python
final['temperature (C)'].fillna(value=294.3, inplace=True)
final['description'].fillna(value='light rain', inplace=True)

#convert kelvin to celcius
final['temperature (C)'] = final['temperature (C)'] - 273.15

final
```

```python
final.to_csv('./datasets/weather_clean.csv')
```
